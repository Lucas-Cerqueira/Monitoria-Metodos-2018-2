### Solução
Seguindo o desenvolvimento abaixo:
$$
\begin{align*}
 -1 &< \sin(n) &< 1 \\
 -3 &< 3\sin(n) &< 3
\end{align*} \\
\frac{-3}{n+\sin(n)} < \frac{3\sin(n)}{n+\sin(n)} < \frac{3}{n+\sin(n)}
$$

Logo:
$$
\lim_{n\to\infty} \frac{-3}{n+\sin(n)} < \lim_{n\to\infty} \frac{3\sin(n)}{n+\sin(n)} < \lim_{n\to\infty} \frac{3}{n+\sin(n)}
$$

Como $\lim_{n\to\infty} \frac{-3}{n+\sin(n)} = 0$ e $\lim_{n\to\infty} \frac{3}{n+\sin(n)} = 0$, pelo Teorema do Confronto:
$$
\lim_{n\to\infty} \frac{3\sin(n)}{n+\sin(n)} = 0
$$

#### Importando bibliotecas

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

#### Definindo as sequências:
$$
    a_n = \lim_{n\to\infty} \frac{-3}{n+\sin(n)} \\
    b_n = \lim_{n\to\infty} \frac{3}{n+\sin(n)} \\
    x_n = \lim_{n\to\infty} \frac{3\sin(n)}{n+\sin(n)} \\
$$
Vimos que $a_n < x_n < b_n$.

```python
# Definindo a sequência de números n
n = np.arange (1.0, 301.0)
# Calculando as sequências
a_n = -3/(n + np.sin(n))
b_n = 3/(n + np.sin(n))
x_n = 3*np.sin(n)/(n + np.sin(n))
```

#### Gráfico das sequências

```python
fig=plt.figure(figsize=(15, 9))
plt.scatter(n, b_n, s=50)
plt.scatter(n, x_n, s=50)
plt.scatter(n, a_n, s=50)
plt.legend(['b_n', 'x_n', 'a_n'])
```
