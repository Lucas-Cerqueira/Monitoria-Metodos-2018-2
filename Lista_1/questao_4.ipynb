### Questão 4

Dados $q, x_1 \in \mathbb{R}$, considere a sequência $(x_n)$ com $n$-ésimo termo $x_n = x_1.q^{n-1}$, para todo $n \in \mathbb{N}$. Seja a sequência $(s_n)$ cujo $n$-ésimo termo é:
$$ s_n = \sum_{k=1}^n x_k, \; \forall n \in \mathbb{N} $$

Disserte sobre a convergência de $(s_n)$.

### Solução

Observe que $s_n$ é a soma de uma P.G. de razão $q$. Para deduzir a expressão de $s_n$, fazemos:
$$
\begin{align*}
s_n &= x_1 + x_1.q + x_1.q^2 + \ldots + x_1.q^{n-1} \\ \\
s_n.q &= x_1.q + x_1.q^2 + x_1.q^3 + \ldots + x_1.q^{n} \\ \\
s_n(1-q) &= x_1(1-q^n) \\ \\
s_n &= \frac{x_1(1-q^n)}{1-q}, \quad q\neq 1
\end{align*}
$$

Observe que o único termo que depende de $n$ é $q^n$:
$$
\begin{align*}
\left\lvert q \right\rvert < 1 &\rightarrow \lim_{} q^n = 0 \\
\left\lvert q \right\rvert > 1 &\rightarrow \lim_{} q^n = \infty
\end{align*}
$$

Com isso, $s_n$ só converge se $\left\lvert q \right\rvert < 1$, levando a:
$$ \lim_{} s_n = \frac{x_1}{1-q}, \quad \left\lvert q \right\rvert < 1 $$

#### Importando bibliotecas

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

#### Definindo uma P.G. com $x_1 = 1$ para diversos valores de $q$.

```python
# Definindo a sequência de números n
n = np.arange (1.0, 21.0) # de 1 (inclusive) até 21 (exclusive)

# Definindo o primeiro termo da P.G.
x_1 = 1
```

##### P.G com $q = 0.5$

```python
q = 0.5
x_n = [x_1 * q**(i-1) for i in n]

plt.scatter(n, x_n)
plt.xlabel('n', fontsize=14)
plt.ylabel('x_n', fontsize=14)
plt.title ('P.G. com q = %.2f' % q, fontsize=14)
```

##### P.G com $q = -0.5$

```python
q = -0.5
x_n = [x_1 * q**(i-1) for i in n]

plt.scatter(n, x_n)
plt.xlabel('n', fontsize=14)
plt.ylabel('x_n', fontsize=14)
plt.title ('P.G. com q = %.2f' % q, fontsize=14)
```

##### P.G com $q = 1.1$

```python
q = 1.1
x_n = [x_1 * q**(i-1) for i in n]

plt.scatter(n, x_n)
plt.xlabel('n', fontsize=14)
plt.ylabel('x_n', fontsize=14)
plt.title ('P.G. com q = %.2f' % q, fontsize=14)
```

##### P.G com $q = -1.1$

```python
q = -1.1
x_n = [x_1 * q**(i-1) for i in n]

plt.scatter(n, x_n)
plt.xlabel('n', fontsize=14)
plt.ylabel('x_n', fontsize=14)
plt.title ('P.G. com q = %.2f' % q, fontsize=14)
```
