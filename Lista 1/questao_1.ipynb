### Questão 1

Prove que $\lim{} \frac{n}{n+1} = 1$

### Solução

Dividindo $\frac{n}{n+1}$ por $n$ em cima e embaixo:

$$\lim{} \frac{n}{n+1} = \lim{} \frac{1}{1+\frac{1}{n}} = 1$$

#### Importando bibliotecas

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

#### Definindo a sequência $n$ e calculando a função desejada

```python
# Definindo a sequência de números n
n = np.arange (1.0, 101.0) # de 1 (inclusive) até 101 (exclusive)
# Calculando a função
fun = n / (n + 1)
```

#### Plot do gráfico

```python
# Plotando n pela função
plt.plot (n, fun, 'o')
# Configurando os eixos
plt.ylabel ("n/(n+1)")
plt.xlabel ("n")
```
